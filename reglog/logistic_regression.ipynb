**Implementación de un Modelo de Clasificación Binaria**

En este notebook, realizaremos el paso a paso del procesamiento de datos a través de un modelo lineal transformado por una función de activación (**sigmoide**)  para predecir la probabilidad de que un cliente vuele de nuevo en la misma aerolínea.


# Definición de variables de Entrada (Features)
El modelo recibe tres características previamente analizadas y limpiadas por el equipo de Ciencia de Datos.


*   Edad
*   Ingreso Mensual
*   Puntuación del vuelo (1 - 5 estrellas)




# 1. Features realistas
Un cliente tiene como valores:


*   Edad = 30 años
*   Ingreso Mensual = $35,000
*   Puntuación del vuelo = 4.5 estrellas

```python
import numpy as np # Para el manejo de arrays

# Features del cliente
x = np.array([30, 35000, 4.5])

# Pesos o weights aprendidos por el modelo de ML (ejemplo)
w = np.array([-0.02, 0.00004, 0.8])
b = -1.5
```

# 2. Modelo lineal
Calculamos el valor de z, que representa la combinación lineal de las variables multiplicadas por sus respectivos pesos (w) más un sesgo o bias (wₙ):

$$z = (w_1 \cdot edad) + (w_2 \cdot ingreso) + (w_3 \cdot puntuación) + b$$

Sustituyendo valores:

$$z = (-0.02 \times 30) + (0.00004 \times 35000) + (0.8 \times 4.5) - 1.5 = 1.1$$

```python
z = np.dot(w, x) + b
print("Salida de modelo lineal: ", z)
```

# 3. Función de activación (sigmoide)

Para convertir el valor ***z*** que puede ir desde -∞ a +∞ en un rango de probabilidad entre 0.0 y 1.0

```python
def sigmoid(z):
  return 1 / (1 + np.exp(-z))

prob = sigmoid(z)
print("Probabilidad de que el cliente vuelva a la aerolínea: ", prob)
```

# 4. Clasificación

Se asigna un umbral en donde si la predicción da arriba de un 0.75 el cliente probablemente volará de nuevo (**Clase 1**), caso contrario no volará (**Clase 0**).

```python
prediction = 1 if prob >= 0.75 else 0

print("Clase predicha: ", prediction)
print("Volará de nuevo, enviando ofertas...") if prediction == 1 else print("No volará de nuevo, enviando encuestas de satisfacción...")
```

# Análisis y ajuste de Features
Explicación de cómo influyen las variables de entrada o features a nuestra predicción final y sus pesos. El objetivo es ajustar los parámetros de la función *predict_return_airline()* para observar cómo cambia la probabilidad de que un cliente regrese.

**1.  Variables de Entrada (Features)**

El modelo considera tres dimensiones del usuario:

  *   Edad
  *   Ingreso Mensual
  *   Puntuación del vuelo (1 - 5 estrellas)

---

**2.  Interpretación de los Pesos (Weights)**

Para entender por qué el modelo toma ciertas decisiones, analizamos el peso asignado a cada feature:

   
**A. Feature Edad** (-0.02 x Edad)

Cada año adicional reduce ligeramente la probabilidad de regreso. Es un impacto sutil negativo:

*   **Usuario de 20 años:** Aporta un -0.4
*   **Usuario de 40 años:** Aporta un -0.8

*A mayor edad, el modelo tiende a ser más conservador al predecir que el cliente regresará.*

---

**B. Feature Ingreso** (-0.00004 x Ingreso)

El ingreso mensual es una variable de gran magnitud (por ejemplo $40,000) por lo que su peso es pequeño para mantener el equilibrio en la escala del modelo.

*   **40,000 x 0.00004:** Aporta un 1.6

*A mayor ingreso mensual aumenta la probabilidad de que el cliente vuelva a viajar con la aerolínea.*

---

**B. Feature Puntuación** (0.8 x Estrellas)

Este es el **predictor más fuerte** del modelo ya que con **3 estrellas**: 3 x 0.8 = 2.4

*A mejor experiencia del vuelo (más estrellas), mayor probabilidad de fidelidad del cliente.*


---

**3.  Interpretación de Bias (sesgo)**

El **bias** representa el punto base del modelo antes de considerar cualquier variable del cliente.


Si todas las features fueran 0 y el bias lo asignamos como -1.5 al finalizar la sigmoide daría: σ(-1.5) = 0.18

Significaría que la Probabilidad base del regreso sería un 18%. Por defecto el modelo cree que la persona no regresará a menos que los features de ingreso mensual y puntuación aporten suficiente evidencia para aumentar esa probabilidad.

```python
def sigmoid_ej(z):
  return 1 / (1 + np.exp(-z))

def predict_return_airline(age, income, score):
  X = np.array([age, income, score])

  w = np.array([-0.02, 0.00004, 0.8])
  b = -1.5

  z = np.dot(w, X) + b
  prob = sigmoid_ej(z)

  prediction = 1 if prob >= 0.75 else 0
  return z, prob, prediction

z, prob, prediction = predict_return_airline(20, 30000, 1.0)
print("Salida de modelo lineal: ", z)
print("Probabilidad de que el cliente vuelva a la aerolínea: ", prob)
print("Clase predicha: ", prediction)
```
